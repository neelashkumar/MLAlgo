# Neural Network Model

# Step 1: Importing Libaries

```python
from keras.models import Sequential#to initialize
from keras.layers import Dense#to add layers
```

# Step 2: Initialize the neural network model

```python
model=Sequential()
x_train.shape
```

# step 3: Add input layer

```python
model.add(Dense(input_dim=11,init="random_uniform",activation="relu",output_dim=20))
```

# step 4: Add hidden layer

```python
model.add(Dense(init="random_uniform",activation="relu",output_dim=15))
```

# step 5: Add output layer

```python
model.add(Dense(init="random_uniform",activation="sigmoid",output_dim=1))
```

```python
#11 neuron in i/p layer
#6 neuron in hidden layer
#1 neuron in output layer
```

# step 6: Compile the NN model

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])#loss='mse' for regression model
```

# step 7: Train the model

```python
model.fit(x_train,y_train,batch_size=32,epochs=100)
```

# step 8: Test the model

```python
y_pred=model.predict(x_test)
```

```python
y_pred#beacuse we used sigmoid function
```

```python
y_test
```

```python
y_pred=y_pred>0.5
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
