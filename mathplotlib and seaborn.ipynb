```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5] #classes-school
```

```python
y=[150,200,50,75,97] #in each class no of students
```

```python
plt.plot(x,y)
plt.xlabel('Class')
plt.ylabel('No of students')
plt.title('School XYZ')
plt.show()
```

```python
y1=[150,200,50,75,97]#school1
y2=[140,225,510,175,971]#school2
plt.plot(x,y1,'r--',label='School1')
plt.plot(x,y2,color='blue',label='school2')
plt.xlabel('Class')
plt.ylabel('No of students')
plt.title('School Data')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y1,label='School1')
plt.scatter(x,y2,label='school2',color='green')
plt.xlabel('Class')
plt.ylabel('No of students')
plt.title('School Data')
plt.legend()
plt.show()
```

```python
plt.bar(x,y1,label='School1',color='red')
plt.bar(x,y2,label='school2',color='green',alpha=0.2)
plt.xlabel('Class')
plt.ylabel('No of students')
plt.title('School Data')
plt.legend()
plt.show()
```

```python
plt.pie(y1,labels=x)
plt.show()
```
