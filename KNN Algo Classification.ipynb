```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
dataset=pd.read_csv(r'C:\Users\test01\Desktop\Social_Network_Ads.csv')
```

```python
dataset
```

```python
dataset=dataset.drop(columns=['User ID'])
```

```python
x=dataset.iloc[:,:-1].values
```

```python
y=dataset.iloc[:,-1].values
```

```python
lb=LabelEncoder()
```

```python
x[:,0]=lb.fit_transform(x[:,0])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# KNN algo

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5,p=2)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
import sklearn.metrics as metrics
```

```python
fpr,tpr,threshold=metrics.roc_curve(y_test,y_pred)
```

```python
roc=metrics.auc(fpr,tpr)
```

```python
roc
```

```python
plt.plot(fpr,tpr,label='AUC=%0.2f' % roc)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC')
plt.legend()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
pd.DataFrame(confusion_matrix(y_test,y_pred),columns=['Prediction-0','Prediction-1'],index=['Actual-0','Actual-1'])
```
