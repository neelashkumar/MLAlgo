```python
video=cv2.VideoCapture(0)
while True:
    data,frame=video.read()#to read each frame
    cv2.imshow("Video",frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):#if q key is pressed then video stops
        break
video.release()
cv2.destroyAllWindows()#to close all windows automatically
```

```python
video=cv2.VideoCapture(0)
while True:
    data,frame=video.read()#to read each frame
    cv2.imshow("Video",frame)
    blur_img=cv2.blur(frame,(10,10))
    cv2.imshow("Blur",blur_img)
    if cv2.waitKey(1) & 0xFF==ord('q'):#if q key is pressed then video stops
        break
video.release()
cv2.destroyAllWindows()#to close all windows automatically
```
